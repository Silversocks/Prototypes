```python
import pysr
import sympy
import numpy as np
from matplotlib import pyplot as plt
from pysr import PySRRegressor
from sklearn.model_selection import train_test_split
import pandas as pd

import re
```

```python
def camel_case_split(str):
    return re.findall(r'[A-Z](?:[a-z]+|[A-Z]*(?=[A-Z]|$))', str)

class SymbRegressor:  # actual regression model
    def __init__(self,niterations=40,maxsize=20,binary_operators=["+", "*"],unary_operators=["cos","exp","sin","inv(x) = 1/x"]):
        self.niterations=niterations #increase for better results
        self.maxsize=maxsize
        self.binary_operators=binary_operators
        self.unary_operators=unary_operators
        self.model=pysr.PySRRegressor(
            maxsize=20,
            niterations=40,  # < Increase me for better results
            binary_operators=self.binary_operators
            unary_operators=self.unary_operators
            extra_sympy_mappings={"inv": lambda x: 1 / x},
            elementwise_loss="loss(prediction, target) = (prediction - target)^2" # maybe later a library of loss functions could be made, for easier access/abstraction reasons
        )
    def fit(self,x,y):
        self.model.fit(x,y)
    def predict(self,x):
        self.model.predict(x)
    def get_features(self,dataset): # helper function to make suitable variable names for each feature
        featurelist=[]
        vardict={}
        if isinstance(dataset,pd.DataFrame):
            featurelist.append(dataset.columns)
        for i in featurelist:
            k=i.split(sep="_") # may need to include support for camelcase and other formats later
            varname=str([j[0] for j in k])
            vardict[i]=varname
            print(vardict)
            return vardict
```
